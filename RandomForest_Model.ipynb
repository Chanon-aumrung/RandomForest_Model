```
import pandas as pd
seed =55 #ใส่เลขอะไรก็ได้เพื่อนำไปเป็นค่า random state
df = pd.read_csv('telco_dataset.csv')
```

```
x =df.iloc[:,1:-1]
y =df.iloc[:,-1:]
x.head()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = seed) 
for temp in [X_train,X_test,y_train,y_test] :
  print(temp.shape)
```

```
from sklearn.ensemble import RandomForestClassifier
random_forest_model = RandomForestClassifier(n_estimators=100,min_samples_split=10,random_state=seed) #n_estimators คือ จำนวน decision tree ใน random forest model
#min_sample_split คือ จำนวนขั้นต่ำของ data ที่จะแตกออกมาในแต่ล่ะ node ของ decision tree
random_forest_model.fit(X_train, y_train)
```

```
from sklearn.metrics import accuracy_score
y_train_pred_rf = random_forest_model.predict(X_train)
y_test_pred_rf = random_forest_model.predict(X_test)

print('Accuracy of model on train set', accuracy_score(y_train,y_train_pred_rf))
print('Accuracy of model on test set', accuracy_score(y_test,y_test_pred_rf))
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix_train = confusion_matrix(y_train,y_train_pred_rf)
print(confusion_matrix_train)

from sklearn.metrics import classification_report
print(classification_report(y_train,y_train_pred_rf))
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix_test = confusion_matrix(y_test,y_test_pred_rf)
print(confusion_matrix_test)

from sklearn.metrics import classification_report
print(classification_report(y_test,y_test_pred_rf))
```

```
#feature important คือ feature ไหนกันแน่ที่มีความสำคัญ
rf_importance = random_forest_model.feature_importances_
rf_importance
```

```
feature_importance_df = pd.DataFrame({'feature' : X_train.columns,
                                      'importance' : rf_importance})
feature_importance_df 
```

```
#total day minutes มีความสำคัญที่สุดเมื่อเทียบกับตัวแปรตัวอื่น เอาประมาณ 3 ตัวแรกที่มัเกาะกลุ่มกัน คือเอาตัวเลขที่มันเกาะกลุ่มกี่ตัวก็ได้กันไปใส่ model คำนวณใหม่อีกทีเพื่อเพิ่มความแม่นยำ
feature_importance_df.sort_values(by = 'importance', ascending=False)
```
